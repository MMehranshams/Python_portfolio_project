### Answering to quentions:

### Q1: Which airline had the highest percentage of delayed or cancelled flights in 2015?

Methodology:
    Filtering flights data for the year 2015.
    Calculating total delayed or cancelled flights per airline.
    Calculating total flights per airline.
    Determining percentage of delayed or cancelled flights per airline.

```python
# Filter flights for the year 2015
flights_2015 = flights_df[flights_df['YEAR'] == 2015]

# Calculate the total number of delayed or cancelled flights per airline
delayed_or_cancelled_flights = flights_2015[(flights_2015['ARRIVAL_DELAY'] > 0) | (flights_2015['CANCELLED'] == 1)]
delayed_or_cancelled_counts = delayed_or_cancelled_flights['AIRLINE'].value_counts()

# Calculate the total number of flights per airline
total_flights_per_airline = flights_2015['AIRLINE'].value_counts()

# Calculate the percentage of delayed or cancelled flights per airline
percentage_delayed_or_cancelled = (delayed_or_cancelled_counts / total_flights_per_airline) * 100

# Find the airline with the highest percentage
highest_percentage_airline = percentage_delayed_or_cancelled.idxmax()
highest_percentage = percentage_delayed_or_cancelled.max()

print("Airline with the highest percentage of delayed or cancelled flights in 2015:")
print("Airline:", highest_percentage_airline)
print("Percentage:", highest_percentage)
```

### Question # 2: Which airports had the most flight cancellations?

```python
# Group flights by origin airport and count cancellations
airport_cancellations = flights_df[flights_df['CANCELLED'] == 1]['ORIGIN_AIRPORT'].value_counts()

# Find the airport with the most cancellations
most_cancellations_airport = airport_cancellations.idxmax()
cancellations_count = airport_cancellations.max()

print("Airport with the most flight cancellations:")
print("Airport:", most_cancellations_airport)
print("Cancellations count:", cancellations_count)
```

### Question 3: Are there any geographical patterns in flight delays? Do certain regions or airports experience more delays than others?

```python
# Merge flights data with airports data to get geographical information
flights_with_airports = flights_df.merge(airports_df, left_on='ORIGIN_AIRPORT', right_on='IATA_CODE')

# Calculate average delay for each airport
average_delay_per_airport = flights_with_airports.groupby('AIRPORT')['ARRIVAL_DELAY'].mean().sort_values(ascending=False)

# Display top 10 airports with highest average delay
print("Top 10 airports with highest average delay:")
print(average_delay_per_airport.head(10))
```

### Question 4: What was the average flight delay for each day of the week? Are weekends or weekdays more prone to delays?

```python
# Calculate average delay for each day of the week
average_delay_per_day = flights_df.groupby('DAY_OF_WEEK')['ARRIVAL_DELAY'].mean()

# Display average delay for each day of the week
print("Average flight delay for each day of the week:")
print(average_delay_per_day)
```

### Visualisation of the percentage of delayed flights

```python
import matplotlib.pyplot as plt

# Calculate percentage of delayed or cancelled flights per airline
percentage_delayed_or_cancelled = (delayed_or_cancelled_counts / total_flights_per_airline) * 100

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(percentage_delayed_or_cancelled, labels=percentage_delayed_or_cancelled.index, autopct='%1.1f%%')
plt.title('Percentage of Delayed or Cancelled Flights per Airline in 2015')
plt.show()
```
